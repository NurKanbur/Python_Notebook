```python
#Boş değerler veri analizimizde hatalı sonuç verebilir.
#Bunu önlemek için çeşitli yollar vardır.
#Bunlardan biri boş verileri silmek, verilerin ortalamasında boş değerleri doldurmak ya da medyanı alınarak boş değerleri doldurmak.
#Burada medyanlarını alarak boş değerleri dolduracağız.
```

```python
rating_median = df["Rating"].median()
rating_median
```

```python
df["Rating"] = df["Rating"].fillna(rating_median)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
df.info()
```

```python
#sayısal olan veri tipleri object olarak tanımlanmış bunları sayısal değişken türleri ile değiştirmeliyiz.
```

```python
df["Reviews"].describe()
```

```python
df["Reviews"] = df["Reviews"].astype("int64")
```

```python
df["Reviews"].describe().round()
```

```python
print(len(df["Size"].unique()))

df["Size"].unique()
```

```python
#Burad "M" ve "K" gibi tip uyuşmazlığı yaratabilecek karakterleri silmemiz gerek.
```

```python
df["Size"] = df["Size"].replace("M", "", regex=True)
```

```python
df["Size"] = df["Size"].replace("k", "", regex=True)
```

```python
df.dropna(inplace=True)
```

```python
df["Size"].unique()
```

```python
size_median = df[df["Size"]!= "Varies with device"]["Size"].astype(float).median()
size_median
```

```python
df["Size"] = df["Size"].replace("Varies with device", size_median)
df.dropna(inplace=True)
```

```python
df.Size = pd.to_numeric(df.Size)
```

```python
df.Size.head()
```

```python
df.Size.describe().round()
```

```python
df["Installs"].unique()
```

```python
df.Installs = df.Installs.apply(lambda x:x.replace("+",""))
df.Installs = df.Installs.apply(lambda x:x.replace(",",""))
df.Installs = df.Installs.apply(lambda x:int(x))
```

```python
df["Installs"].unique()
```

```python
df["Price"].unique()
```

```python
df.Price = df.Price.apply(lambda x:x.replace("$",""))
df.Installs = df.Installs.apply(lambda x:float(x))
```

```python
df["Price"].unique()
```

```python
len(df["Genres"].unique())
```

```python
df["Genres"].head(10)
```

```python
df["Genres"] = df["Genres"].str.split(";").str[0]
```

```python
len(df["Genres"].unique())
```

```python
df["Genres"].unique()
```

```python
df["Genres"].value_counts()
```

```python
df["Genres"] = df["Genres"].replace("Music & Audio","Music")
```

```python
df["Last_Updated"].head()
```

```python
df["Last_Updated"] = pd.to_datetime(df["Last_Updated"])
```
