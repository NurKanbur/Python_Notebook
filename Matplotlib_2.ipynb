```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,30)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
(yeniFigure, yeniEksen) = plt.subplots()
```

```python
(yeniFigure, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi2,color = "#3A95A8")
#alpha saydamlık vermek için kullanılıyor "alpha=0.3"
yeniEksen.plot(numpyDizisi2,numpyDizisi1, color = "#C96F23")
```

```python
(yeniFigure, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi2 + 2, "b", linewidth = 4.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 3, "g",linewidth = 4.0)

yeniEksen.plot(numpyDizisi1,numpyDizisi2 + 200, color = "#3A95A8",linewidth = 4.0,linestyle =":",marker = "o",markerfacecolor = "red")
```
