## ÖZELLEŞTİRME İŞLEMLERİ

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b*-")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"p")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.subplot(1,2,2) # şu an ikinci grafiği çiziyoruz.
plt.plot(numpyDizisi2,numpyDizisi1,"b*-")
```

```python
yeniFigure = plt.figure()
```

```python
yeniFigure = plt.figure()
figureAxes = yeniFigure.add_axes([0.4,0.4,0.6,0.6]) # boyutlandırma için kullanılıyor.
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("grafik başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.3,0.4,0.3,0.3]) #küçük tablonun konumu için önemli rol oynar.
eksen1.plot(numpyDizisi1,numpyDizisi2,"b")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana grafik başlığı")

eksen2.plot(numpyDizisi2,numpyDizisi1,"r")
eksen2.set_xlabel("X ekseni küçük")
eksen2.set_ylabel("Y ekseni küçük")
eksen2.set_title("küçük grafik başlığı")
```

## GÖRSEL GELİŞTİRMELER

```python
yeniFigure2 = plt.figure(dpi = 200)
#dpi sonradan da belirtilebilir.
#savefig("yeniFigure.png",dpi = 200) diyerek kayıt edebiliriz.
eksen3 = yeniFigure2.add_axes([0.1,0.1,0.9,0.9])
eksen3.plot(numpyDizisi1,numpyDizisi2 ** 2,label = "numpyDizisi ** 2")
eksen3.plot(numpyDizisi1,numpyDizisi2 ** 3,label = "numpyDizisi ** 3")
eksen3.legend() #yerini değiştirebiliriz.
eksen3.legend(loc = 6) # bu şekilde legend konumu belirleyebiliriz.
```
